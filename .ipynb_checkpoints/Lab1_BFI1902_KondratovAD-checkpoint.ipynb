### Задание №3

```python
import random
from queue import PriorityQueue

# Сортировка выбором
# Проходим по элементам списка, меняя их местами с наименьшими элементами, которые идут после них по списку

def sel_sort(array):
    for i in range(len(array) - 1):
        m = i
        j = i + 1
        while j < len(array):
            if array[j] < array[m]:
                m = j
            j = j + 1
        array[i], array[m] = array[m], array[i]
    return array

# Сортировка вставкой
# Проходим по элементам списка, двигая все предыдущие элементы, которые больше текущего элемента вправо
# После чего переносим текущий элемент в подходящую для него позицию

def insertion_sort(array):
    for i in range(1, len(array)):
        key = array[i]
        j = i-1
        while j >=0 and key < array[j] :
            array[j+1] = array[j]
            j -= 1
        array[j+1] = key 
    return array

# Сортировка обменом (пузырьком)
# Проходим по списку множество раз, меняя попарно элементы в зависимости от их значений

def bubble_sort(array):
    for i in range(len(array) - 1):
        for j in range(len(array) - i - 1):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
    return array

# Сортировка Шелла

def shell_sort(array):
    inc = len(array)
    while inc:
        for i, num in enumerate(array):
            while i >= inc and array[i - inc] > num:
                array[i] = array[i - inc]
                i -= inc
            array[i] = num
        if inc == 2:
            inc = 1 
        else:
            inc = int(inc * 5.0 / 11)
    return array

# Турнирная сортировка

def tourney_sort(array):
    sorted_array = []
    pq = PriorityQueue()
    for i, num in enumerate(array):
        pq.put(num)
    while not pq.empty():
        sorted_array.append(pq.get()) 
    return sorted_array

# Быстрая сортировка
# Из списка выбираем случайный опорный элемент
# Разбиваем список на 3 списка (список элементов больших, меньших и равных опорному)
# Для списков больших и меньших чисел повторить сортировку рекурсивно

def quick_sort(array):
    if len(array) <= 1:
        return array
    else:
        q = random.choice(array)
        smaller_array = []
        bigger_array = []
        equal_array = []
        for n in array:
            if n < q:
                smaller_array.append(n)
            elif n > q:
                bigger_array.append(n)
            else:
                equal_array.append(n)
        return quick_sort(smaller_array) + equal_array + quick_sort(bigger_array)

# Пирамидальная сортировка

def heapify(array, array_length, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < array_length and array[i] < array[l]:
        largest = l
    if r < array_length and array[largest] < array[r]:
        largest = r
    if largest != i:
        array[i],array[largest] = array[largest],array[i]
        heapify(array, array_length, largest)

def heap_sort(array):
    for i in range(len(array), -1, -1):
        heapify(array, len(array), i)
    for i in range(len(array) - 1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify(array, i, 0)
    return array


m = int(input() or "50")
n = int(input() or "50")
min_limit = int(input() or "-250")
max_limit = int(input() or "1006")

new_array1 = [[random.randint(min_limit, max_limit) for row in range(0,n)] for row in range(0,m)]
for i in range(m):
    print(sel_sort(new_array1[i]))
print("\n")
    
new_array2 = [[random.randint(min_limit, max_limit) for row in range(0,n)] for row in range(0,m)]
for i in range(m):
    print(insertion_sort(new_array2[i]))
print("\n")

new_array3 = [[random.randint(min_limit, max_limit) for row in range(0,n)] for row in range(0,m)]
for i in range(m):
    print(bubble_sort(new_array3[i]))
print("\n")

new_array4 = [[random.randint(min_limit, max_limit) for row in range(0,n)] for row in range(0,m)]
for i in range(m):
    print(shell_sort(new_array4[i]))
print("\n")

new_array5 = [[random.randint(min_limit, max_limit) for row in range(0,n)] for row in range(0,m)]
for i in range(m):
    print(tourney_sort(new_array5[i]))
print("\n")

new_array6 = [[random.randint(min_limit, max_limit) for row in range(0,n)] for row in range(0,m)]
for i in range(m):
    print(quick_sort(new_array6[i]))
print("\n")

new_array7 = [[random.randint(min_limit, max_limit) for row in range(0,n)] for row in range(0,m)]
for i in range(m):
    print(heap_sort(new_array7[i]))
print("\n")
```
